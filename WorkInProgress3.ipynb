```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def one_step(weights, pre_syn, post_syn):
    dt=0.01
    tau=10
    new_post_syn=post_syn+dt/10*(-post_syn+np.dot(weights,sigmoid(pre_syn)))
    return new_post_syn
```

```python
xs=[0]
for j in range(0,10000):
    xs.append(one_step(weights,input_patterns[:,j],xs[-1]))
```
